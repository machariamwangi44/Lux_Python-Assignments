```python
#Example: Create  a fi;e sample.txt and write hello world into the file
with open("sample.txt","w")as file:
    file.write("Hello world")
```

```python
#Read the content of the above file
with open("sample.txt","r") as file:
    content=file.read()
    print(content)
```
